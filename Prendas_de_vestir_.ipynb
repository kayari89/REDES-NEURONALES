```python
import keras
# cargue el conjumto de datos Fashion  MNIST
fmnist = keras.datasets.fashion_mnist
```

```python
# Cargue la división de entrenamiento y prueba del conjunto de datos Fashion MNIST
(training_images, training_labels), (test_images, test_labels) = fmnist.load_data()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# puedes poner aqui entre 0 y 59999
index = 1
# imprime la etiqueta  y la imagen.
np.set_printoptions(linewidth=320)
print(f'label:{training_labels[index]}')
print(f'image:\n{training_images[index]}')
# visualiza la imagen
plt.imshow(training_images[index])
```

```python
# normalizar los valores de pixeles del tren y probar las imagenes
training_images = training_images / 255.0
```

```python
# construir el modelo de clasificacion
model = keras.models.Sequential([
    keras.Input(shape=(28, 28)), # Define la forma de entrada explícitamente
    keras.layers.Flatten(),       # Flatten ya no necesita input_shape
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
# compilar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
# Entrenar el modelo
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(training_images, training_labels, epochs=5)
```

```python
# Entrenar el modelo
history = model.fit(training_images, training_labels, epochs=10)
```

```python
import pandas as pd
# graficar el historial de entrenamiento:
pd.DataFrame(history.history).plot(grid=True)
```

```python
#Evaluar el modelo en el conjunto de entrenamiento
loss, accuracy =model.evaluate(training_images, training_labels)
print("Perdida en el conjunto de entrenamiento:",loss)
print("precision en el conjunto de entrenamiento:", accuracy)
```

```python
# Evaluar el modelo con datos no vistos
loss,accuracy =model.evaluate(test_images, test_labels)
print("perdida en el conjunto de prueba:", loss)
print("precision en el conjunto de prueba:", accuracy)
```

```python
#predict
index = 1
print(f'label:{test_labels[index]}')
classification = model.predict(test_images[index:index+1])
print(f'Classification:\n {classification.reshape(-1,1)}')
```
