```python
# 🌿 Kayari - Sistema de Reconexión v4
import csv
from datetime import datetime

def test_reconexion():

    print("🌿 Bienvenida a Kayari - Test de Reconexión Consciente 🌿")

    nombre = input("¿Cuál es tu nombre? ")
    print(f"\nHola {nombre}, vamos a explorar tu estado emocional actual.\n")

    infusiones = {
        "estresada": {
            "nombre": "Infusión Liberación",
            "proposito": "Soltar tensiones y recuperar calma.",
            "ritual": "Respira profundo 3 veces antes de beberla."
        },
        "ansiosa": {
            "nombre": "Infusión Serenidad",
            "proposito": "Regular el sistema nervioso y volver al presente.",
            "ritual": "Toma la infusión en silencio durante 5 minutos."
        },
        "cansada": {
            "nombre": "Infusión Renacer",
            "proposito": "Recuperar energía y vitalidad emocional.",
            "ritual": "Agradece en voz alta antes del primer sorbo."
        },
        "desmotivada": {
            "nombre": "Infusión Fuego Interno",
            "proposito": "Reactivar tu propósito y pasión.",
            "ritual": "Escribe una intención mientras la bebes."
        },
        "sin claridad": {
            "nombre": "Infusión Amanecer del Alma",
            "proposito": "Activar claridad mental y dirección.",
            "ritual": "Bébela mirando hacia la luz natural."
        },
        "desconectada": {
            "nombre": "Infusión Raíz",
            "proposito": "Volver a tu centro y estabilidad.",
            "ritual": "Coloca una mano en tu corazón al beberla."
        },
        "en paz": {
            "nombre": "Infusión Equilibrio",
            "proposito": "Mantener armonía interior.",
            "ritual": "Compártela con alguien que amas."
        }
    }

    estado = input(
        "¿Cómo te sientes hoy? "
        "(estresada / ansiosa / cansada / desmotivada / sin claridad / desconectada / en paz): "
    ).lower()

    if estado in infusiones:
        seleccion = infusiones[estado]
        print("\n✨ Recomendación personalizada Kayari ✨")
        print(f"🌿 Infusión: {seleccion['nombre']}")
        print(f"🎯 Propósito: {seleccion['proposito']}")
        print(f"🔮 Ritual sugerido: {seleccion['ritual']}")

        # Guardar en archivo CSV
        fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        with open("registro_kayari.csv", mode="a", newline="", encoding="utf-8") as archivo:
            escritor = csv.writer(archivo)
            escritor.writerow([nombre, estado, seleccion["nombre"], fecha])

        print("\n📁 Tu registro ha sido guardado en el sistema Kayari.")

    else:
        print("\nEscucha tu intuición y elige la infusión que más resuene contigo.")

# Ejecutar
test_reconexion()

```

```python
import pandas as pd

# Leer archivo
df = pd.read_csv("registro_kayari.csv", header=None)

# Nombrar columnas
df.columns = ["Nombre", "Emocion", "Infusion", "Fecha"]

# Mostrar registros
print("📊 Historial Emocional Kayari\n")
display(df)

# Conteo de emociones
print("\n📈 Conteo de estados emocionales:\n")
print(df["Emocion"].value_counts())
```

```python
import matplotlib.pyplot as plt

conteo = df["Emocion"].value_counts()

plt.figure()
conteo.plot(kind="bar")
plt.title("Distribución Emocional Kayari")
plt.xlabel("Emoción")
plt.ylabel("Frecuencia")
plt.show()
```
